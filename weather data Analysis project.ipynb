```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
path = r'C:\Users\vasud\OneDrive\Documents\weather_data.csv'
df = pd.read_csv(path)
#df.to_csv('weather_data.csv',index=False)
print(df)
```

```python
# Display the first few rows of the dataset
print(df.head())

# Display basic information about the dataset
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values or drop rows/columns with missing values
df = df.dropna()  # For simplicity, we'll drop rows with missing values
```

```python
# Display the column names
print(df.columns)
```

```python
# Distribution of maximum temperature
plt.figure(figsize=(10, 8))
sns.histplot(df['MaxTemp'], bins=45, kde=True)
plt.title('Maximum Temperature Distribution')
plt.xlabel('MaxTemp')
plt.ylabel('Frequency')
plt.show()

# Relationship between maximum temperature and humidity at 3pm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MaxTemp', y='Humidity3pm', data=df)
plt.title('MaxTemp vs Humidity3pm')
plt.xlabel('MaxTemp')
plt.ylabel('Humidity3pm')
plt.show()
```

```python
# Average maximum temperature by wind direction at 3pm
avg_max_temp_by_wind_dir = df.groupby('WindDir3pm')['MaxTemp'].mean()

plt.figure(figsize=(10, 6))
avg_max_temp_by_wind_dir.plot(kind='bar')
plt.title('Average Maximum Temperature by Wind Direction at 3pm')
plt.xlabel('Wind Direction at 3pm')
plt.ylabel('Average MaxTemp')
plt.show()
```

```python
# Save the cleaned DataFrame to a new CSV file
df.to_csv('weather_data_cleaned.csv', index=False)
print("Data saved to 'weather_data_cleaned.csv'")
```
